                                                        Applying the KNN 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('User_Data.csv')
df
```

```python
dummy_gender = pd.get_dummies(df['Gender'], prefix='Gender')

# Concatenate the dummy variables with the original DataFrame
df = pd.concat([df, dummy_gender], axis=1)

# Drop the original "Gender" column if needed
df.drop('Gender', axis=1, inplace=True)
df
```

```python
df['Gender_Male'] = df['Gender_Male'].replace({True:1, False: 0})
df['Gender_Female'] = df['Gender_Female'].replace({True:1, False: 0})
df
```

```python
column_order = ['Age','EstimatedSalary','Gender_Male','Gender_Female','Purchased']
df=df[column_order]
df
```

```python
# Training and Test data using
from sklearn.model_selection import train_test_split
train,test = train_test_split(df,test_size = 0.2)
```

```python
train
```

```python
test
```

```python
# KNN using sklearn
# Importing Knn algorithm from sklearn.neighbors
from sklearn.neighbors import KNeighborsClassifier as KNC
```

```python
# for 3 nearest neighbours
neigh = KNC(n_neighbors= 3)
neigh
```

```python
# Fitting with training data
neigh.fit(train.iloc[:,0:4],train.iloc[:,4])
```

```python
train_pred=neigh.predict(train.iloc[:,0:4])
train_pred
```

```python
# train accuracy
train_acc = np.mean((neigh.predict(train.iloc[:,0:4]))==(train.iloc[:,4])) # 94 %
print(train_acc*100)
```

```python
# test accuracy
test_acc = np.mean(neigh.predict(test.iloc[:,0:4])==test.iloc[:,4]) # 100%
print(test_acc*100)
```

```python
# for 5 nearest neighbours
neigh = KNC(n_neighbors=5)
neigh
```

```python
# fitting with training data
neigh.fit(train.iloc[:,0:4],train.iloc[:,4])
```

```python
# train accuracy
train_acc = np.mean(neigh.predict(train.iloc[:,0:4])==train.iloc[:,4])
train_acc*100
```

```python
# test accuracy
test_acc = np.mean(neigh.predict(test.iloc[:,0:4])==test.iloc[:,4])
test_acc*100
```
