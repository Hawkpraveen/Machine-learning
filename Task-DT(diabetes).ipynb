```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("Diabetes_RF.csv")
data
```

```python
data.columns
```

```python
colnames = list(data.columns)
colnames
```

```python
predictors = colnames[:8]
predictors
```

```python
target = colnames[8]
target
```

```python
# splitting data into training and testing data set
from sklearn.model_selection import train_test_split
train,test = train_test_split(data,test_size=0.2)
```

```python
train
```

```python
test
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion= 'entropy')
model.fit(train[predictors],train[target])
```

```python
# Accuracy = train
y_pred = model.predict(train[predictors])
y_pred
```

```python
pd.crosstab(train[target],y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(train[target],y_pred)
```

```python
y_pred = model.predict(test[predictors])
y_pred
```

```python
pd.crosstab(test[target],y_pred)
```

```python
accuracy_score(test[target],y_pred)
```
